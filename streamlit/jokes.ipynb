```python
import pandas as pd
import re
```

```python
data = pd.read_csv('../data/jokes.csv')
data
```

```python
# Función para extraer el año de la columna text
def extract_year(text):
    match = re.search(r"\b(19|20)\d{2}\b", text)  # Busca años desde 1900 hasta 2099
    if match:
        return match.group(0)
    else:
        return None
```

```python
# Aplicar la extracción del año y crear la columna 'edicion'
data["year"] = data["event_name"].apply(extract_year)  # Extraer el año
data["edicion"] = "Viña " + data["year"]  # Crear la columna 'edicion'
```

```python
data.info()
```

```python
# Convertir la columna 'year' a tipo numérico
data['year'] = pd.to_numeric(data['year'], errors='coerce')

# Crear la columna 'decada' con el formato deseado
def obtener_decada(year):
    if pd.isna(year):
        return None
    decada = (year // 10) * 10
    if decada == 2000:
        return '00'
    elif decada == 2010:
        return '10'
    elif decada < 2000:
        return str(decada)[-2:]  # Tomamos los últimos dos dígitos
    else:
        return str(decada)[-2:]  # Mismo formato para futuras décadas

data['decada'] = data['year'].apply(obtener_decada)
data
```

```python
jokes_df = data[["show_name", "edicion", "decada", "text"]]  # Seleccionar las columnas deseadas
```

```python
jokes_df.to_csv('jokes_df.csv', index=False)  # Guardar el archivo limpio
jokes_df
```
