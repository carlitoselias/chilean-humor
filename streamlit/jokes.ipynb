```python
import pandas as pd
import re
```

```python
data = pd.read_csv('../data/jokes.csv')
data
```

```python
# Función para extraer el año de la columna text
def extract_year(text):
    match = re.search(r"\b(19|20)\d{2}\b", text)  # Busca años desde 1900 hasta 2099
    if match:
        return match.group(0)
    else:
        return None
```

```python
# Aplicar la extracción del año y crear la columna 'edicion'
data["year"] = data["event_name"].apply(extract_year)  # Extraer el año
data["edicion"] = "Viña " + data["year"]  # Crear la columna 'edicion'
```

```python
jokes_df = data[["show_name", "text", "edicion"]]  # Seleccionar las columnas deseadas
```

```python
jokes_df.to_csv('jokes_df.csv', index=False)  # Guardar el archivo limpio
```
